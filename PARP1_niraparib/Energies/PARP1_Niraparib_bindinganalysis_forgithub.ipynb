```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
DDG_data = pd.read_csv("./Energies_PARP1.txt")
```

```python
## Convert mutation names to a more readable format
DDG_data["start"] = DDG_data["Mutation"].str[:1]
DDG_data["end"] = DDG_data["Mutation"].str[2:]
DDG_data["Mutationname"] = DDG_data["start"] + DDG_data["end"]
```

```python
DDG_data
```

```python
g = sns.barplot(y = DDG_data.sort_values(by = "DDG").head(10)["DDG"]
                , x = DDG_data.sort_values(by = "DDG").head(10)["Mutationname"], edgecolor = "k",
               palette = sns.color_palette("Set3"))
g.tick_params(top=True, labeltop=True, bottom=False, labelbottom=False)

g.set_ylim(-2, 0)
g.set_xlabel("")
g.set_xticklabels(g.get_xticklabels(), size = 10)
g.set_yticklabels(g.get_yticklabels(), size = 12)
g.set_ylabel("Binding DDG (kcal/mol)", size = 14)
sns.despine(top = False, bottom = True)

plt.savefig("Niraparib_binding_lowest10.pdf", bbox_inches="tight")

plt.show()
```

```python
g = sns.barplot(y = DDG_data.sort_values(by = "DDG").tail(10)["DDG"]
                , x = DDG_data.sort_values(by = "DDG").tail(10)["Mutationname"], edgecolor = "k",
               palette = sns.color_palette("Set3"))

g.set_xlabel("")
g.set_xticklabels(g.get_xticklabels(), size = 10)
g.set_yticklabels(g.get_yticklabels(), size = 12)
g.set_ylabel("Binding DDG (kcal/mol)", size = 14)
sns.despine()

plt.savefig("Niraparib_binding_highest10.pdf", bbox_inches="tight")

plt.show()
```
